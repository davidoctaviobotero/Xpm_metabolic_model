# Xam Metabolic Model BiGG

```python
import cobra
import os
from os.path import join

data_dir = "/Users/davidbotero/Dropbox (UCSD SBRG)/Xam_Multistrain_Recon/Metabolic_model/BiGG_Model/"
xam_model = cobra.io.read_sbml_model(join(data_dir, "Xam_BiGG_minmal_media.xml"))
```

```python
print (len(xam_model.reactions))
print (len(xam_model.metabolites))
print (len(xam_model.genes))
```

### Reactions

```python
print("Reactions")
print("---------")
for x in xam_model.reactions:
    print("%s : %s" % (x.id, x.reaction))
```

```python
print("Names Reactions")
print("---------")
for x in xam_model.reactions:
    print("%s : %s" % (x.id, x.name))
```

### Transport and Exchange reactions

```python
count=0
for reactions in xam_model.reactions:
    for x in reactions.metabolites:
        if (x.compartment == 'e'):
            print("%s : %s" % (reactions.id, reactions.reaction))            
            count += 1
print(count)
```

### Metabolites

```python
print("")
print("Metabolites")
print("-----------")
for x in xam_model.metabolites:
    print('%9s : %s, %s, %s' % (x.id, x.name, x.charge, x.compartment))
```

### Genes

```python
print("")
print("Genes")
print("-----")
for x in xam_model.genes:
    associated_ids = (i.id for i in x.reactions)
    print("%s\t%s" %
          (x.id, ",".join(associated_ids)))
```
