```python
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')

from time import time

import cobra
import os
from os.path import join
from cobra.flux_analysis import calculate_phenotype_phase_plane

data_dir = "/Users/davidbotero/Dropbox (UCSD SBRG)/Xam_Multistrain_Recon/Metabolic_model/BiGG_Model/"

xam_model = cobra.io.read_sbml_model(join(data_dir, "Xam_BiGG_V1.0.txt"))

%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
import matplotlib.pyplot as plt
```

### Exchange reactions

```python
count = 0
for reactions in xam_model.reactions:
    if len(reactions.metabolites) == 1:
        print("%s : %s" % (reactions.name, reactions.reaction))
        count += 1
print(count)
```

### Phase Plane

```python
data = calculate_phenotype_phase_plane(xam_model, "EX_cpd00007_e0", "EX_cpd00027_e0")
data.plot_matplotlib();
plt.savefig('Phase_Plane_Gluc-O2_Xam.svg', format='svg', dpi=1200)
plt.savefig('Phase_Plane_Gluc-O2_Xam.png', format='png', dpi=1200)
```

- EX_cpd00027_e0: D-glucose exchange</p>
