```python
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')

from time import time

import cobra
import os
from os.path import join
from cobra.flux_analysis import calculate_phenotype_phase_plane

data_dir = "/Users/david/Desktop/Xam_Multistrain_Recon/Metabolic_model/BiGG_Model/"

xam_model = cobra.io.load_json_model(join(data_dir, "Xam_Biomass&Xanthan(completed_mar08)_model.json"))

%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')
import matplotlib.pyplot as plt
```

### Exchange reactions

```python
count = 0
for reactions in xam_model.reactions:
    if len(reactions.metabolites) == 1:
        print("%s : %s %s %s" % (reactions.id, reactions.reaction, reactions.lower_bound, reactions.upper_bound))
        count += 1
print(count)
```
