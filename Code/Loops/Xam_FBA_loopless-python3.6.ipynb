```python
import cobra
from cobra.flux_analysis.loopless import add_loopless, loopless_solution
from cobra.flux_analysis import pfba

import os
from os.path import join

import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
from time import time

import pandas as pd
pd.options.display.max_rows = 300
```

```python
cobra_config = cobra.Configuration()
cobra_config.solver = "glpk"
```

```python
xam_model = cobra.io.load_json_model("Xam_BiGG_full_media.json")
```

### Original model

```python
print (len(xam_model.reactions))
print (len(xam_model.metabolites))
print (len(xam_model.genes))
```

### FBA

```python
xam_model.objective = "Biomass"
FBA_sol = xam_model.optimize()
print(FBA_sol)
```

### Loopless FBA

```python
from cobra.flux_analysis.loopless import add_loopless, loopless_solution
from cobra.flux_analysis import pfba
```

```python
FBA_loopless_sol = loopless_solution(xam_model)
#print(FBA_loopless_sol)
```

```python
df = pd.DataFrame(dict(loopless=FBA_loopless_sol.fluxes, nominal=FBA_sol.fluxes))
```

```python
df.plot.scatter(x='loopless', y='nominal')
```

### FVA Loopless

```python
from cobra.flux_analysis import flux_variability_analysis
```

```python
FVA_xam = flux_variability_analysis(xam_model, xam_model.reactions[:])
```
