```python
import os
from os.path import join

import pandas
pandas.options.display.max_rows = 300

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import cobra
```

```python
cobra_config = cobra.Configuration()
cobra_config.solver = "glpk"

data_dir = "../Models/"
xam_model = cobra.io.load_json_model(join(data_dir, "Xam_BiGG_full_media.json"))
```

### Original model

```python
print (len(xam_model.reactions))
print (len(xam_model.metabolites))
print (len(xam_model.genes))
```

```python
c=0
for i in [i.id for i in xam_model.reactions if len(i.metabolites) == 1]:
    reaction = xam_model.reactions.get_by_id(i)
    print('%s\t%i\t%i' % (reaction.id, reaction.lower_bound, reaction.upper_bound))
    c+=1
print(c)
```

### FBA

```python
xam_model.objective = "Biomass"
FBA_sol = xam_model.optimize()
print (FBA_sol)
```

```python
FBA_results = FBA_sol.fluxes
FBA_results = FBA_results.rename_axis("Reaction")
FBA_results = FBA_results.rename("Flux")
print(FBA_results)
```

```python
fba_df_sort = FBA_results.sort_values(axis=0)
plt.figure()
fba_df_sort.plot.line()
```

### FVA

```python
fva_results = cobra.flux_analysis.flux_variability_analysis(xam_model, xam_model.reactions, fraction_of_optimum=1)
#fva_df = pandas.DataFrame.from_dict(fva_results).T.round(5)
fva_results.index.name = 'reaction'
fva_results.sort_index(kind='heapsort', inplace=True)
print(fva_results)
```

```python
fva_df_sort = fva_results.sort_values(by=['maximum', 'minimum'])
print(fva_df_sort)
```

```python
plt.figure(dpi=400)
fva_df_sort.plot.box()
#plt.savefig('FVA_Xam_box_80.svg', format='svg', dpi=1200)
#plt.savefig('FVA_Xam_box_80.png', format='png', dpi=1200)
fva_df_sort.plot.line()
#plt.savefig('FVA_Xam_plot_80.svg', format='svg', dpi=1200)
#plt.savefig('FVA_Xam_plot_80.png', format='png', dpi=1200)
```

### Fluxes over  a Threshold

```python
import numpy

fva_df_sort_equal = fva_df_sort[(fva_df_sort["maximum"] > 900) | (fva_df_sort["minimum"] < -900)]
print(fva_df_sort_equal)
```

```python
print(len(fva_df_sort_equal))
```

### Exporting S matrix to an array - only FVA fluxes

```python
import numpy as np
import re

#Exporting S_matrix to array formating
#The type of array to construct. if dense, return a standard numpy.array,
# dok, or lil will construct a sparse array using scipy of the corresponding type
# and DataFrame will give a pandas DataFrame with metabolite indices and reaction columns

S_matrix_sparse = cobra.util.array.create_stoichiometric_matrix(xam_model, array_type="dok")
S_matrix_dense = cobra.util.array.create_stoichiometric_matrix(xam_model, array_type="dense")
S_matrix_df = cobra.util.array.create_stoichiometric_matrix(xam_model, array_type="DataFrame")
```

```python
S_matrix_df_T = S_matrix_df.T
S_matrix_df.T
```

```python
S_matrix_FVA = fva_df_sort_equal.join(S_matrix_df_T, lsuffix='_caller', rsuffix='_other')
S_matrix_FVA = S_matrix_FVA.drop(['minimum', 'maximum'], axis=1)
S_matrix_FVA_T = S_matrix_FVA.T
#S_matrix_FVA_T = S_matrix_FVA_T.set_index('reaction')
S_matrix_FVA_T
```

```python
# Deleting cero rows (metabolites) of S-Matrix

S_matrix_FVA_reduced = S_matrix_FVA_T[(S_matrix_FVA_T.T != 0).any()]
S_matrix_FVA_reduced
```

```python
S_matrix_FVA_reduced.to_csv("FVA_matrix_reduced_full_media.csv")
```
